```python
import pandas as pd
```

```python
json_df= pd.read_json('movie.json')
```

```python
json_df.head()
```

```python
actors = json_df['actors'].tolist()

unwrapped = []
```

```python
for l in actors:
    for m in l:
        unwrapped.append(m)
unwrapped
```

```python
j_df = json_df
```

```python
del j_df['rating'],j_df['year'],j_df['users_rating'], j_df['votes'], j_df['metascore'], j_df['img_url'], j_df['countries'] 
del j_df['languages'],  j_df['genre'], j_df['tagline'], j_df['description'], j_df['directors'], j_df['runtime'], j_df['imdb_url']
```

```python
j_df.head()
```

```python
j_titles = j_df['title'].tolist()
j_actors = j_df['actors'].tolist()
```

```python
copied_movie_list = []
```

```python
counter = 0
for a in j_actors:
    for n in range(0, len(a)):
        copied_movie_list.append(j_titles[counter])
    counter+=1
```

```python
copied_movie_list
```

```python
actor_df = pd.DataFrame(
    {'title': copied_movie_list,
     'performer': unwrapped
    })
```

```python
actor_df.head(50)
```

```python
actor_df.to_csv('actors.csv', index = False)
```
